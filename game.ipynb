```python
import random
```

```python
class BettingGame:
    def __init__(self):
        self.coins = 100
        self.win_prob = 0.6
        self.last_outcome = "NONE"

    def get_state(self):
        if self.coins < 50:
            capital = "LOW"
        elif self.coins < 150:
            capital = "MEDIUM"
        else:
            capital = "HIGH"

        return (capital, self.last_outcome)

    def step(self, bet: int):
        if bet > self.coins:
            bet = self.coins

        win = random.random() < self.win_prob
        old_coins = self.coins

        if win:
            self.coins += bet
            self.last_outcome = "WIN"
        else:
            self.coins -= bet
            self.last_outcome = "LOSS"

        reward = self.coins - old_coins
        return self.get_state(), reward
```

```python
game = BettingGame()

print(game.get_state())   # start

state, reward = game.step(5)
print(state, reward)

state, reward = game.step(2)
print(state, reward)

state, reward = game.step(1)
print(state, reward)


state, reward = game.step(0)
print(state, reward)
```
